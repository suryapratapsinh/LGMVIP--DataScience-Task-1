# Seprate input and output columns

```python
data = df.values
x=data[:,0:4]
y=data[:,4]
```

# Seprate data into training and testing

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
print(x_test)
```

# Model selection

# SVM

```python
from sklearn.svm import SVC

mo_svc = SVC()
mo_svc.fit(x_train,y_train)
```

```python
prediction_svc = mo_svc.predict(x_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction_svc))
```

# Logistic Regration

```python
from sklearn.linear_model import LogisticRegression

mo_lr = LogisticRegression()
mo_lr.fit(x_train,y_train)

```

```python
prediction_lr = mo_lr.predict(x_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction_lr))
```

# Decision tree

```python
from sklearn.tree  import DecisionTreeClassifier

mo_dt = DecisionTreeClassifier()
mo_dt.fit(x_train,y_train)
```

```python
prediction_dt = mo_dt.predict(x_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction_dt))
```

# KNN

```python
from sklearn.neighbors  import KNeighborsClassifier

mo_knn = KNeighborsClassifier()
mo_knn.fit(x_train,y_train)
```

```python
prediction_knn = mo_knn.predict(x_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction_knn))
```

# Compare actual o/p and predicted o/p

```python
for i in range(len(prediction_knn)):
  print(y_test[i],prediction_knn[i])   
```
